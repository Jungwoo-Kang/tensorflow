# relu and its variant

$$\begin{array}{llllllll}
\mbox{tf.nn.relu}(x)&=&
\left\{\begin{array}{llllllll}
x&\mbox{for}\ x>0\\
0&\mbox{otherwise}
\end{array}\right.\\
\mbox{tf.nn.leaky_relu}(x)&=&
\left\{\begin{array}{llllllll}
x&\mbox{for}\ x>0\\
\alpha*x&\mbox{otherwise}
\end{array}\right.\\
\mbox{tf.nn.elu}(x)&=&
\left\{\begin{array}{llllllll}
x&\mbox{for}\ x>0\\
\alpha*(e^x-1)&\mbox{otherwise}
\end{array}\right.\\
\mbox{tf.nn.selu}(x)&=&
\left\{\begin{array}{llllllll}
\mbox{scale}*x&\mbox{for}\ x>0\\
\mbox{scale}*\alpha*(e^x-1)&\mbox{otherwise}
\end{array}\right.\\
\mbox{tf.nn.relu6}(x)&=&
\left\{\begin{array}{llllllll}
6&\mbox{for}\ x\ge 6\\
x&\mbox{for}\ 0\le x\le 6\\
0&\mbox{for}\ x<0
\end{array}\right.\\
\end{array}$$

# relu

```python
%matplotlib inline
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
tf.reset_default_graph()
x = np.linspace(-10,10)
y = tf.nn.relu(x)
with tf.Session() as sess:
    y_run = sess.run(y)
plt.plot(x,y_run)
plt.plot(x,np.zeros_like(x),'b', alpha=0.3)
plt.plot([0,0],[np.min(y_run),np.max(y_run)],'b', alpha=0.3)
plt.show()
```

# sigmoid and its variant

$$\begin{array}{llllllll}
\mbox{tf.nn.sigmoid}(x)&=&
\frac{1}{e^{-x}+1}\\
\displaystyle
\mbox{tf.nn.tanh}(x)&=&
2*\mbox{tf.nn.sigmoid}(x)-1\\
\mbox{tf.keras.activations.hard_sigmoid}(x)&=&
\mbox{tf.clip_by_value}(0.2x + 0.5, 0, 1)
\end{array}$$

# sigmoid

```python
%matplotlib inline
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
tf.reset_default_graph()
x = np.linspace(-10,10)
y = tf.nn.sigmoid(x)
with tf.Session() as sess:
    y_run = sess.run(y)
plt.plot(x,y_run)
plt.plot(x,np.zeros_like(x),'b', alpha=0.3)
plt.plot([0,0],[np.min(y_run),np.max(y_run)],'b', alpha=0.3)
plt.show()
```

# hard_sigmoid

```python
%matplotlib inline
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
tf.reset_default_graph()
x = np.linspace(-10,10)
x = tf.convert_to_tensor(x, tf.float32) # caution
y = tf.keras.activations.hard_sigmoid(x)
with tf.Session() as sess:
    x_run, y_run = sess.run([x,y])
plt.plot(x_run,y_run)
plt.plot(x_run,np.zeros_like(x_run),'b', alpha=0.3)
plt.plot([0,0],[np.min(y_run),np.max(y_run)],'b', alpha=0.3)
plt.show()
```

# softmax and others

$$\begin{array}{llllllll}
\mbox{tf.nn.softmax}(x)&=&
\frac{e^x}{\mbox{np.sum}\left(e^x\right)}\\
\displaystyle
\mbox{tf.nn.softplus}(x)&=&
\log(e^x+1)&\approx&|x|^+&\mbox{for large $|x|$}\\
\mbox{tf.nn.softsign}(x)&=&
\frac{x}{|x|+1}&\approx&sgn(x)&\mbox{for large $|x|$}\\
\end{array}$$

# softmax

```python
import tensorflow as tf
import numpy as np
tf.reset_default_graph()
x = np.array([[0.,1.,2.]])
y = tf.nn.softmax(x) 
with tf.Session() as sess:
    y_run = sess.run(y)
print(y_run)
x_temp = np.exp(x)
y_run_np = x_temp / np.sum(x_temp) 
print(y_run_np)
```

# softplus

```python
import tensorflow as tf
import numpy as np
tf.reset_default_graph()
x = np.array([0.,1.,2.])
y = tf.nn.softplus(x)
with tf.Session() as sess:
    y_run = sess.run(y)
print(y_run)
y_run_np = np.log(np.exp(x)+1)
print(y_run_np)
```

# softsign

```python
import tensorflow as tf
import numpy as np
tf.reset_default_graph()
x = np.array([0.,1.,2.])
y = tf.nn.softsign(x)
with tf.Session() as sess:
    y_run = sess.run(y)
print(y_run)
y_run_np = x / (np.abs(x)+1)
print(y_run_np)
```
